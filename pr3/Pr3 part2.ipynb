```python
def manual_mean(values):
    total = sum(values)
    count = len(values)
    return total / count


mean_sepal_length = manual_mean(df['SepalLengthCm'])
print(f"The manually calculated mean of sepal length is: {mean_sepal_length}")
mean_sepal_width = manual_mean(df['SepalWidthCm'])
print(f"The manually calculated mean of sepal length is: {mean_sepal_width}")
mean_petal_length = manual_mean(df['PetalLengthCm'])
print(f"The manually calculated mean of sepal length is: {mean_petal_length}")
mean_petal_width = manual_mean(df['PetalWidthCm'])
print(f"The manually calculated mean of sepal length is: {mean_petal_width}")

```

```python
def manual_median(values):
    sorted_values = sorted(values)


    n = len(sorted_values)

    if n % 2 == 1:
        median = sorted_values[n // 2]
    else:

        middle1 = sorted_values[n // 2 - 1]
        middle2 = sorted_values[n // 2]
        median = (middle1 + middle2) / 2

    return median


median_sepal_length = manual_median(df['SepalLengthCm'])
print(f"The manually calculated median of sepalLengthcm is: {median_sepal_length}")

median_sepal_width = manual_median(df['SepalWidthCm'])
print(f"The manually calculated median of sepalLengthcm is: {median_sepal_width}")

median_petal_length = manual_median(df['PetalLengthCm'])
print(f"The manually calculated median of sepalLengthcm is: {median_petal_length}")


median_petal_width = manual_median(df['PetalWidthCm'])
print(f"The manually calculated median of sepalLengthcm is: {median_petal_width}")
```

```python
import math
def manual_standard_deviation(values):
    mean_value = manual_mean(values)  # Calculate the mean
    squared_diff = [(x - mean_value) ** 2 for x in values]  # Calculate squared differences
    variance = sum(squared_diff) / len(values)  # Variance is the average of squared differences
    std_deviation = math.sqrt(variance)
    return std_deviation
std_dev_sepal_length = manual_standard_deviation(df['SepalLengthCm'])
print(f"The manually calculated standard deviation of sepalLengthcm is: {std_dev_sepal_length}")
std_dev_sepal_width = manual_standard_deviation(df['SepalWidthCm'])
print(f"The manually calculated standard deviation of sepalLengthcm is: {std_dev_sepal_width}")
std_dev_petal_length = manual_standard_deviation(df['PetalLengthCm'])
print(f"The manually calculated standard deviation of sepalLengthcm is: {std_dev_petal_length}")
std_dev_sepal_width= manual_standard_deviation(df['PetalWidthCm'])
print(f"The manually calculated standard deviation of sepalLengthcm is: {std_dev_sepal_width}")
```

```python
def manual_skewness(values):
    mean_value = manual_mean(values)
    std_dev = manual_standard_deviation(values)
    median=manual_median(values)
    skewness = (3 * (mean_value - median)) / std_dev
    return skewness
skew_SepalLen= manual_skewness(df['SepalLengthCm'])
print(f"Skewness of SepalLen: {skew_SepalLen}")
skew_SepalWid= manual_skewness(df['SepalWidthCm'])
print(f"Skewness of SepalWid: {skew_SepalLen}")
skew_PetallLen= manual_skewness(df['PetalLengthCm'])
print(f"Skewness of PetallLen: {skew_PetallLen}")
skew_PetalWid= manual_skewness(df['PetalWidthCm'])
print(f"Skewness of PetalWid: {skew_PetalWid}")
```
