```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load the Titanic dataset
titanic = sns.load_dataset("titanic")

# Set style
sns.set_style("whitegrid")

# 1. Age Distribution Histogram
plt.figure(figsize=(6,4))
sns.histplot(titanic['age'].dropna(), kde=True, bins=30)
plt.title("Distribution of Age")
plt.show()

# 2. Age vs Fare Scatter Plot
sns.jointplot(x="age", y="fare", data=titanic, kind="scatter")
plt.show()

# 3. Rug Plot of Fare
plt.figure(figsize=(6,4))
sns.rugplot(titanic['fare'], height=0.1)
plt.title("Rug Plot of Fare")
plt.show()

# 4. Survival Rate by Gender
plt.figure(figsize=(6,4))
sns.barplot(x="sex", y="survived", data=titanic)
plt.title("Survival Rate by Gender")
plt.show()

# 5. Passenger Count by Class
plt.figure(figsize=(6,4))
sns.countplot(x="class", data=titanic)
plt.title("Passenger Count by Class")
plt.show()

# 6. Age Distribution (Strip Plot)
plt.figure(figsize=(6,4))
sns.stripplot(x="class", y="age", data=titanic, jitter=True, hue="sex", dodge=True)
plt.title("Age Distribution (Strip Plot)")
plt.show()

# 7. Age Distribution (Swarm Plot)
plt.figure(figsize=(6,4))
sns.swarmplot(x="class", y="age", data=titanic, hue="sex")
plt.title("Age Distribution (Swarm Plot)")
plt.show()

# 8. Correlation Heatmap
plt.figure(figsize=(10,6))

# Select only numeric columns
numeric_titanic = titanic.select_dtypes(include=["number"])

# Compute and plot correlation heatmap
sns.heatmap(numeric_titanic.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# 9. Cluster Map
plt.figure(figsize=(10, 6))

# Select only numeric columns
numeric_titanic = titanic.select_dtypes(include=["number"])

# Compute and plot the cluster map
sns.clustermap(numeric_titanic.corr(), cmap="coolwarm", standard_scale=1)
plt.suptitle("Cluster Map of Titanic Features")  # Note: clustermap uses suptitle instead of title
plt.show()

# 10. Fare Histogram
plt.figure(figsize=(10, 6))
sns.histplot(titanic["fare"], bins=30, kde=True, color="skyblue")
plt.xlabel("Fare Price")
plt.title("Fare Distribution")
plt.show()
```
